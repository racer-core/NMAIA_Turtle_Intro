Computational science is very similar to the science you perform in science class. in that you're goal is to gain a new understanding or insight through experimentation. However, in computational science or scientific computing, you'll be using your computer to carry out your experiment, perform analysis of the results, and repeat if necessary.

Let's recall the steps of the Scientific Method:
1. Ask a question. This can be as simple as "what happens if I do this?" and in our case, we can use our computers to test the case.
2. Research the topic. When exploring with our computer, we often need to identify the tools or libraries we'll need to perform our experiment, we need to research the background for the topic of the experiment.
3. Construct a hypothesis. This is our educated guess about the outcome of our experiment and our own reasoning as to why it happens. Often we'll be looking at variables in the experiment and trying to predict their effect on the result.
4. Experimentation to test our hypothesis. Setup and perform a test, evaluate the result, and repeat the experiment if necessary.

```python
# Turtle Expiriment #3: Utilize a for loop to simplify our code (adjust the  # of sides)
initializeTurtle() # This line resets the turtle and canvas

side_length = 170
number_of_sides = 6
angle = 360/number_of_sides #90 degrees

# Now using a "for loop" to simplify our code
for i in range(number_of_sides): # [1 2 3 4 5 6]
  print("Loop number: " + str(i)) # This line will print the loop number below the graphic
  forward(side_length)
  right(angle)
```

```python
# Turtle Prompt: One ring is good, so multiple rings must be great!
# This prompt shows you how to draw more than one ring by changing the turtle angle between drawing operations
# Questions:
#           Could we use another "for loop" to draw a lot more times?
#           Could we write a custom function for the drawing loop?

initializeTurtle()  # This line resets the turtle and canvas
side_length = 25
number_of_sides = 3
angle = 360/number_of_sides
draw_times = 4
draw_angle = 360/draw_times

# Adding a vector of colors for our turtle to use
# More colors can be found here: https://www.w3schools.com/colors/colors_names.asp
color_list = ['mediumblue', 'BlueViolet', 'Coral', 'DarkMagenta','SeaGreen', 'FireBrick', 'GhostWhite',  'Teal']

# Creating a custom function for drawing an N-sided shape
def custom_shape(number_of_sides, side_length, color_list):
  # To repeat an operation N-times, we'll use a "for" loop
  for i in range(number_of_sides):
    color(color_list[i % len(color_list)])
    forward(side_length)
    right(int(360/number_of_sides))

# Calling our custom function to draw a N-sided shape
custom_shape(36, side_length, color_list)
```
